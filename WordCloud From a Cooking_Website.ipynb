### WEB SCRAPPING

```python
import requests # this will connect with HTTPS and needs to have internet connectivity
from bs4 import BeautifulSoup # Web Scrapping Tools
```

```python
#Asssing url that we need to scrape the data
url = "https://kitchenofdebjani.com/2015/10/20-bengali-dishes-to-try-before-you-die/"
r = requests.get(url)
```

```python
#Saving the text part of webpage
html_doc = r.text
```

```python
# Creating beautiful soup element
txt01 = BeautifulSoup(html_doc)
txt01
```

```python
webpg_title=txt01.title
webpg_title
```

```python
webpg_text=txt01.get_text()
#webpg_text
```

### Removing PUNCTUATION

```python
import string
#string.punctuation
```

```python
## function to remove puntuation
def remove_punct(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    return text_nopunct
```

```python
#text after removing punctuation
webpg_text_nopunct=remove_punct(webpg_text)
```

### Tokenization

```python
import re
#function for Tokenizing
def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens
```

```python
webpg_text_tokenized=tokenize(webpg_text_nopunct)
webpg_text_tokenized
```

### Removing Stopwords

```python
#import nltk # NLTK - Natural Language Tool Kit is library to perform analysis on text
#nltk.download('stopwords') # to be downloaded first time
```

```python
from nltk.corpus import stopwords
```

```python
stp=stopwords.words()
#stp
```

```python
#### Function to find stopwords
def remove_stopwords(tokenized_list):

    text = [word for word in tokenized_list if word not in stp]
    
    return text
```

```python
#small list running
tkn_words = ['i', 'am', 'going', 'to', 'go', 'to', 'the', 'store', 'and', 'park']

words_after_tkn=[word for word in tkn_words if word not in stp]
words_after_tkn
```

```python
webpg_text_nostopwrds=[word for word in webpg_text_tokenized if word not in stp]
#webpg_text_nostopwrds
```

### WORDCLOUD

```python
comment_words= ' '
for item in webpg_text_nostopwrds: 
        comment_words = comment_words + item+ ' '
        
#comment_words
```

```python
# importing all necessery modules 
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
```

```python
wordcloud = WordCloud(width = 800, 
                      height = 800,   
                      background_color ='white', 
                      #stopwords = stp, 
                      min_font_size = 10)

wordcloud.generate(comment_words)
```

```python
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 


plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```
